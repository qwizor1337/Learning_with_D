```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop('pet', axis = 1)
y = df['pet']

x_tr, x_ts, y_tr, y_ts = train_test_split(x, y, random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
lr.fit(x_tr, y_tr)

y_pred = lr.predict(x_ts)
```

```python
import numpy as np
def mape(y_true, y_pred): 
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mape(y_ts, y_pred) # MAPE хорош тем, что показывает  отклонение в %
```

#### ДЗ. Тренировочная 2002:2013 Тестовая 2014:2015, проверить модель

### Добавление фичей

```python
df.head()
```

```python
d_180_list = [i for i in df if '180D' in i]
d_60_list = [i for i in df if '60D' in i]
```

```python
for i in range(len(d_180_list)):
    col1 = d_180_list[i]
    col2 = d_60_list[i]
    new_column = col1 + '/' + col2 #'D_180/D_60'
    df[new_column] = df[col1] / df[col2]
```

```python
df.head()
```

```python
df = df.replace([np.inf, -np.inf], np.nan).dropna()
x2 = df.drop('pet', axis = 1)
y2 = df['pet']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x2, y2, random_state = 1)
```

```python
x2
```

```python
errors = []
```

```python
for param in params:
    huber = HuberRegressor(alpha = param['alpha'], epsilon= param['epsilon'])
    huber.fit(x_train, y_train)
    
    pred = huber.predict(x_test)
    
    cv_mape = mape(y_test, pred)
    errors.append({'mape':cv_mape, 'alpha':param['alpha'], 'epsilon':param['epsilon']})
```

```python
errors = pd.DataFrame.from_records(errors)    # попробовать запустить в коллаб, должны быть разные значения
```

```python
errors # результат улучшился с 26 до 9 путем добавления новых переменных
```

```python
errors.iloc[errors['mape'].idxmin()]
```
