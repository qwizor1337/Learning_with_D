## Разбиение на Train и Test

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= .33, random_state = 1) # random state делает так, что всегда одно и то же на выходе
```

### Первая модель машинного обучения

```python
from sklearn.ensemble import RandomForestClassifier

rfc = RandomForestClassifier(max_depth = 3, n_estimators = 1000)
rfc.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

pred = rfc.predict(x_test)
accuracy_score (pred, y_test)
```
