### Основы работы с библиотекой yfinance

```python
import pandas as pd
import numpy as np
import yfinance as yf
```

```python
tickers = ['IEF', 'GLD', 'SPY', 'QQQ']
daily = {}
```

```python
for i in range(len(tickers)):
    data = yf.Ticker(tickers[i]).history(period="max")
    daily[tickers[i]] = data
```

### Своя стратегию торговли (как выбирать лучший тикер и сколько его держать)

```python
for ticker in tickers:
    print(len(daily[ticker]))
# мы располагаем информацией за разные промежутки времени, нужно подготовить данные таким образом, чтобы везде была информация за одинаковый период
```

```python
def data_cutting(data, tickers, tables_len):
    min_len = min(tables_len)
    for ticker in tickers:
        data[ticker] = data[ticker][::-1].iloc[:min_len,:]
        data[ticker] = data[ticker][1:][::-1]

def tickers_download(tickers, dictionary):
    tables_len = []
    for i in range(len(tickers)):
        data = yf.Ticker(tickers[i]).history(period="max")
        daily[tickers[i]] = data
    
        data['pct_change'] = data[::-1]['Close'].pct_change()
        length = len(data)
        tables_len.append(length)
    
    data_cutting(daily, tickers, tables_len)
```
