```python
transac.purchdate = pd.to_datetime(transac['purchdate'])
transac.groupby([pd.Grouper(level='purchdate', freq='M'), 
            pd.Grouper(level='mcc'), 
            pd.Grouper(level='cnum')]).sum()
```

```python
transac
```

```python
transac1 = transac.copy()
```

```python
col = transac1['mcc'].unique()
for name in col:
  transac1[name] = 0
  transac1[name] = transac1.loc[transac1.mcc == name]['amount']
```

```python
transac1.set_index('purchdate', inplace = True)
```

```python
transac1.fillna(0, inplace = True)
transac1.resample('M').sum()
```

```python
transac1.groupby(['purchdate', 'cnum']).sum()
```

```python
transacts =  transac1.groupby('cnum').sum()
transacts.drop('amount', axis = 1, inplace = True)
transacts
```

```python
buys_dict = {}
for i in range(len(transacts)):
  a = list(transacts.iloc[i])
  b = max(a)
  c = a.index(b)
  buys_dict[i] = c
```
