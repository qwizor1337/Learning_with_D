```python
trans_df = pd.read_csv(path+'tr100k.csv', index_col = 0).copy()
cl_list_df = cl_lst.copy()
trans_df.drop(['mrchcity', 'mrchname'], axis = 1, inplace = True)
cl_list_df.gender.replace({'M':0, 'F': 1}, inplace = True)
cl_list_df.married_.replace({'not_married':0, 'married': 1}, inplace = True)
cl_list_df.residenttype.replace({'R':0, 'N': 1}, inplace = True)
cl_list_df.rename(columns = {'cnum_':'cnum'}, inplace = True)
```

```python
df = pd.merge(trans_df, cl_list_df, on = 'cnum')
df.drop('purchdate', axis = 1, inplace = True)
df.set_index('cnum', inplace = True)
df['amount'] = df['amount'].astype(np.int64)
df
```

```python
x = df.iloc[:,1:]
y = df.iloc[:,0:1]
```
